```python
sentments = {}

def sent_one_message(a,b):
    res = client.message(b)['entities']
    sentments[a] = res

def mutiple_thread(idx,text):
    jobs =[]

    for a,b in zip(idx,text):
        try:
            #thread class represents an activity that is running in a separate thread of control
            thread = threading.Thread(target=sent_one_message(a,b)) 
            jobs.append(thread)
            thread.start()

        except Exception:
            continue

    for j in jobs:
            j.join()
```
